```python
import pandas as pd 
import numpy as np 
```

```python
people = {
    'first' : ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last' : ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email' : ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age' : ['33', '55', '63', '36', None, None, 'Missing']

}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna()
df.dropna(axis='index', how='any')
# custom missing value still in the dataframe
# set axis='index' we say want to drop in a values when our rows has missing values when its
# if we set to columnsthen it would instead columns they had missing value

# how='any' it will drop any missing values
# how='any' it will drop if all the values missing
```

```python
df.dropna(axis='index', how='all')
```

```python
df.dropna(axis='index', how='any', subset=['email'])
# when we want to do it in one columns
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email'])
# email OR last name we do not have
```

```python
df.dropna(axis='index', how='any', subset=['last', 'email'])
# email AND last name we do not have
```

```python
# replace custom value with np.nan
df.replace('NA', np.nan)
df.replace('Missing', np.nan)
```

```python
df.isna()
# see if are there Nan value
```

```python
# MOSTLY USE IN NUMERIC VALUE
# sometimes we replace NaN value with certain value
df.fillna('Missing')
```

```python
df.dtypes
# we can see age column is object
type(np.nan) # float
```

```python
df['age'].mean() # this is str we need float type data

# if column has no nan value 'astype' method works but here it gives ERROR
# because np.nan is float
df['age'] = df['age'].astype(int)
```

```python
df.replace('Missing', 0, inplace=True)
df.replace('NA', 0, inplace=True)
df.dropna()
df['age'] = df['age'].astype(float)
df.dtypes
```

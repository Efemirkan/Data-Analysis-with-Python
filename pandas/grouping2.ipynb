##### FIGURE OUT HOW MANY PEOPLE IN EACH COUNTRY KNOW TO USE PYTHON


```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
# sum can work in Boolen object 1 aND 0 then we can see total of one 's knows Python
```

```python
country_grp = df.groupby(['Country'])

```

```python
country_grp['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# has no attribute 'str' therefore we need to apply method with lambda to call
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

##### WHAT % PEOPLE IN EACH COUNTRY KNOW TO USE PYTHON

```python
country_respondests = df['Country'].value_counts()
country_respondests
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# create new database with this informations
python_df = pd.concat([country_respondests, country_uses_python], axis='columns', sort=False)
# df.reset_index(inplace=True) delete index
python_df
```

```python
# change columns names
python_df.rename(columns={'count': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
# add percentage column
python_df['PercKnowsPython'] = (python_df['NumKnowsPython']/ python_df['NumRespondents']) * 100
```

```python
# sort by largest percentage
python_df.sort_values(by='PercKnowsPython', ascending=False, inplace=True)
```
